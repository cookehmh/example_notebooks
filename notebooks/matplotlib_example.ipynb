```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure()
x = np.arange(1,10)
y = np.arange(11,20)
plt.plot(x,y)
```

```python
plt.figure()

x = np.arange(1,10)
y = np.arange(11,20)
plt.plot(x, y, color = 'red', marker = 'o', label = 'dataset 1')

x = np.arange(11,1,-1)
y = np.arange(5,15)
plt.plot(x, y, color = 'b', ls = '', marker = 'v', label = 'dataset 2')

plt.legend(fontsize = 15, frameon = True)
plt.ylabel('y axis', fontsize = 15)
plt.xlabel('x axis', fontsize = 15)
plt.tick_params(labelsize = 15)
```

```python
data = {'a': np.arange(100),
        'c': np.random.randint(0, 100, 100),
        'd': np.random.randn(100)}
data['b'] = data['a'] + 10 * np.random.randn(100)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```
